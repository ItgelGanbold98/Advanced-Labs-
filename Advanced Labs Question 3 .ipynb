(a) Use the numpy.random.rand() function to generate two arrays containing the
x and y coordinates of 100 virtual raindrops. Note: numpy.random.rand(N)
generates N random numbers from a uniform distribution between 0 and 1, or
just generates one random number if N is not specified.

(b) For visualisation plot the raindrops and draw the quarter circle (take care with
the aspect ratio!). Note: keep the plotting separate from the calculation below.

```python
#generate 100 random points
N=100
x,y = np.random.rand(N), np.random.rand(N) 

#define the quarter circle function
x_circle = np.linspace(0,1,200)
y_circle = np.sqrt(1-x_circle**2)
#draw the circle and plot the points 
plt.scatter(x,y, c='red')
plt.plot(x_circle, y_circle, c = 'blue')
plt.axis('scaled')
plt.axhline(c='grey')
plt.axvline(c='grey')
plt.title("The ratio of a quarter circle to the square to estimate \u03C0")
plt.grid();
```

(c) Determine a logical test to decide whether each raindrop falls within the quarter
circle and use it to count up how many raindrops fall within the quarter circle.

(d) Use the ratio of the number of raindrops in the circle to the total number in the
square to estimate π.

```python
#make a function that tests if a raindrop is either inside or outside the circle
# it is assumed that the probability that any point lands precisely on the circle is negligible
def counter(a,b):
    i = 0 
    for j in range(len(a)):
        if b[j] <= np.sqrt(1-a[j]**2):
            i = i+1
    return i

#the ratio of the total rain drops and the ones inside the circle should approximate to 4/pi as the number of 
#rain drops tend towards infinity
print(f"The approximate value for \u03C0 using N= 100 is {4*counter(x,y)/N}")
```

(e) Run your program a few times and investigate how the estimate of π changes.
Hint: you could make a function which takes N and returns π to make running
it multiple times and for different values of N easier.

```python

#make a program that calculates approximate value for pi for given N values

def pi(start,final,interval):
    M = np.arange(start,final, interval)
    z=0
    pi_est = []
    while z < len(M):
        x,y = np.random.rand(M[z]), np.random.rand(M[z])
        k = 4*(counter(x,y))/M[z]
        z+=1
        pi_est.append(k)
    return pi_est 

N = np.arange(100,10000,100)
plt.plot(N,pi(100,10000,100), label='Approximate value for \u03C0')
plt.grid()
y1 = N*0+np.pi
plt.plot(N,y1,c='r',label='Actual value of \u03C0')
plt.legend()
plt.xlabel('N')
plt.title('How the estimate of \u03C0 changes with N')
print("As N increases, the estimated value of \u03C0 gets closer to the actual value.");
```
