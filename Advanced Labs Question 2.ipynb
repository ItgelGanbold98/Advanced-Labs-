## Q2 
The equation of a cosine wave of amplitude A and frequency $\nu$, and initial phase angle
ϕ as a function of time is:
## $y(t)=A\cos(2\pi\nu t+\phi)$

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

(a) Plot a cosine function with frequency 10 Hz, amplitude 1.0 (arbitrary unit) and phase angle 30◦
in the range t = 0 to 2 seconds (use 500 points).

```python
 

def y(A,nu,t,phi):     #define y(t)
    return (A*np.cos(2*np.pi*nu*t+phi))

def deg_to_rad(i):
    return i * np.pi/180


phi_30 = deg_to_rad(30)

t = np.linspace(0,2,500)
y1 = y(1,10,t,phi_30)
plt.plot(t,y1)
plt.grid()
plt.ylabel("y(t)")
plt.xlabel("t")
plt.title("Frequency of 10");
```

(b) On a single graph plot in different colours two cosine functions, one of frequency 10
Hz, amplitude 1.0 and phase angle 30◦ and the other of frequency 8 Hz, amplitude
0.8 and phase angle 60◦

```python
y2 = y(0.8,8,t, deg_to_rad(60))
plt.plot(t, y2, label = 'A=0.8, nu=8, phi=60', color='green')
plt.plot(t,y1,  label = 'A=1, nu=10, phi=30', color='blue')
plt.title('Frequency of 10Hz and Frequency of 8Hz overlapped')
plt.grid()
plt.ylabel("y(t)")
plt.xlabel("t");
```

(c) Determine and plot the resulting wave that occurs when the two waves from (2b)
interfere.

```python
y3 = y1+y2
plt.plot(t,y3, color='black')
plt.title(label = 'Interference of the two waves')
plt.grid()
plt.ylabel("y(t)")
plt.xlabel("t");
```

(d) Plot the frequency spectrum (amplitude -vs- frequency) determined by a fast
Fourier transform (FFT) of the interference pattern from exercise (2c):

i. Use np.fft.fft() which takes a numpy array as argument and returns the
fast Fourier transform (FFT) of it.

ii. The FFT algorithm returns an array of complex amplitudes with the second
half of the array being a mirror copy of the first half - hence you will want to
create a new array that contains the np.abs() of the first half of the array
returned by the FFT function.

iii. The frequencies of the elements of the array from (2(d)ii) are: 0, $f_\min$, $2f_\min$,
$3f_\min$ . . .$N/2\times f_\min$, where $f_\min$ = 1/T where T is the maximum time and N is
the number of points in the input array to the FFT function.
Use subplots to create a second plot to ’zoom’ in on the region of interest. Is the
resulting graph as expected?


```python
graph, axis = plt.subplots(2,1) #make 2 sub plots
plt.tight_layout() 
fourier = np.fft.fft(y3)/len(y3) 
fourier = fourier[range(int(len(y3)/2))]*2     #split in 2
freq = np.arange(250) / 2

axis[0].plot(freq,np.abs(fourier))
axis[0].set_title("Fast Fourier Transform")
axis[0].grid()

axis[0].set_ylabel("Amplitude")
axis[1].axis([0,20,0,1.05])
axis[1].plot(freq_,np.abs(fourier))
axis[1].grid()
axis[1].set_xlabel("Frequency (Hz)")
axis[1].set_ylabel("Amplitude");

#output shows frequencies of 8Hz and 10Hz, same as the original waves
```
